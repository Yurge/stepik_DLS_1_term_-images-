```python
# ---------------------------------------------- LogisticRegression ----------------------------------------------

# Функция для поиска параметров с помощью GridSearchCV
def manual_search_params(X_train=X_train, y_train=y_train):
    # инициализируем класс
    params = {'model__C': [60.]}        # [10., 30., 40., 50., 60., 70., 100.]

    model_log_reg = Pipeline(steps=[
        ('scaler', StandardScaler()),
        ('model', LogisticRegression(max_iter=1000, solver='saga', random_state=42))
        ])

    # Поиск лучших параметров
    gridsearch = GridSearchCV(
        model_log_reg,
        param_grid=params,
        cv=5,
        scoring='roc_auc',
        n_jobs=-1
    )

    # обучаем модель на train
    log_reg = gridsearch.fit(X_train, y_train)

    print(f'best score LogisticRegression: {log_reg.best_score_}')
    print(f'best params LogisticRegression: {log_reg.best_params_}')

    return log_reg


log_reg = manual_search_params()

y_predict_proba = log_reg.predict_proba(X_test)[:, 1]

submission['Churn'] = y_predict_proba
submission.to_csv('submission_logreg_manual.csv', index=False)      # ROC-AUC вашего решения равен 0.84532


# Функция для поиска параметров с помощью LogisticRegressionCV
def log_regression_auto_search(X_train=X_train, y_train=y_train, X_test=X_test):
    c_values = [100., 10., 1., 0.1, 0.001, 0.01]
    logit_searcher = LogisticRegressionCV(Cs=c_values, max_iter=1000, cv=5, n_jobs=-1, scoring='roc_auc')
    logit_searcher.fit(X_train, y_train)

    y_predict_prob = logit_searcher.predict_proba(X_test)[:, 1]

    return y_predict_prob


y_pred_prob_log_reg_auto = log_regression_auto_search()
submission['Churn'] = y_pred_prob_log_reg_auto
submission.to_csv('submission_logreg_auto.csv', index=False)        # ROC-AUC вашего решения равен 0.84531
```

```python
# ---------------------------------------------- RandomForestClassifier ----------------------------------------------

# Функция для поиска параметров с помощью GridSearchCV
def search_params(X_train=X_train, y_train=y_train):

    tree_params_rf = {
		'n_estimators': [27],       # range(15, 20, 25)
		'criterion': ['entropy'],  	# ['gini', 'entropy']
		'max_depth': [11],  			# range(7, 8, 9)
		'max_features': [3],  		# range(2, 3, 4)
		'min_samples_split': [70],  # range(70, 71, 72)
		'min_samples_leaf': [16]  	# range(15, 16, 17)
	}

    clf_rf = RandomForestClassifier(random_state=42)

    gridsearch = GridSearchCV(
		estimator=clf_rf,
		param_grid=tree_params_rf,
		cv=5,
		scoring='roc_auc'
	)

    # обучаем модель на train
    random_forest = gridsearch.fit(X_train, y_train)

    print(f'best score RandomForestClassifier: {random_forest.best_score_}')
    print(f'best params RandomForestClassifier: {random_forest.best_params_}')

    return random_forest


random_forest = search_params()

y_pred_proba = random_forest.predict_proba(X_test)[:, 1]

submission['Churn'] = y_pred_proba
submission.to_csv('submission_random_forest.csv', index=False)      # ROC-AUC вашего решения равен 0.85046
```

```python

X_train_nn = torch.FloatTensor(X_train.values)
y_train_nn = torch.FloatTensor(y_train.values)
X_test_nn = torch.FloatTensor(X_test.values)

print(X_train_nn.shape, '\n', y_train_nn.shape, '\n', X_test_nn.shape, '\n')
```

```python
#
import numpy as np
import torch
from torch import nn


class NN_churn(nn.Module):
    def __init__(self, n_hidden_neurons):
        super(NN_churn, self).__init__()
        func = nn.ReLU6()
        self.fc1 = nn.Linear(19, n_hidden_neurons)
        self.act1 = func
        self.fc2 = nn.Linear(n_hidden_neurons, 5)
        self.act2 = func
        self.fc4 = nn.Linear(5, 1)
        self.act4 = func
        self.sm = nn.Softmax(dim=1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.fc2(x)
        x = self.act2(x)
        x = self.fc4(x)
        x = self.act4(x)
        return x

    def inference(self, x):
        x = self.forward(x)
        x = self.sm(x)
        return x


n_hidden_neurons = 150
my_net = NN_churn(n_hidden_neurons)

loss = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(my_net.parameters(), lr=0.0001)

batch_size = 100
mse = []
for epoch in range(325):
    l = []
    order = np.random.permutation(len(X_train_nn))

    for start_index in range(0, len(X_train_nn), batch_size):
        optimizer.zero_grad()

        batch_indexes = order[start_index:start_index + batch_size]

        if len(batch_indexes) < batch_size:
            break

        x_batch = X_train_nn[batch_indexes]
        y_batch = y_train_nn[batch_indexes]

        predict = my_net(x_batch)
        predict = torch.reshape(predict, (batch_size,))

        loss_val = loss(predict, y_batch)
        l.append(loss_val.item())

        loss_val.backward()
        optimizer.step()

    mse.append(np.mean(l))

print(len(mse))

y_pred_proba = my_net.forward(X_test_nn).detach().numpy()
#

```
